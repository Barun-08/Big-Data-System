```python
import pyspark
from pyspark.sql import SparkSession
```

```python
spark = SparkSession.builder \
        .appName("PySpark DataFrame Basics") \
        .getOrCreate()

spark
```

```python
df = spark.read.csv('/content/product_info.csv', header=True, inferSchema=True)

df.show()
```

```python
df.printSchema()
```

```python
df.describe().show()
```

```python
df.summary().show()
```

```python
df.select("product_name").show()
```

```python
df.select("product_name", "rating").show()
```

```python
df = df.withColumn("loves_count_after_increasing_10_count", df["loves_count"] + 10)

df.show()
```

```python
df = df.drop("loves_count_after_increasing_10_count")

df.show()
```

```python
df = df.withColumnRenamed("product_name", "Name")

df.show()
```
